```python
import pandas as pd
import plotly.express as px 

#importar a base de dados (jah baixada e na mesma pasta do .py)
tabela = pd.read_csv("telecom_users.csv")
display(tabela)

```

```python
for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna,color="Churn")
    grafico.show()
```

```python
# excluir uma coluna
tabela = tabela.drop("Unnamed: 0", axis=1) #axis é eixo , 1 é coluna e 0 é linha . se quiser deletar varias colunas, por o nome delas em []. depois de axis=(

#analizar se o python tah analizando a tabela no formato correto
#sera que há alguma coluna vazia?

tabela["total gasto"] = pd.to_numeric(tabela["total gastos"], errors="coerce")

#excluir linhas e colunas vazias
tabela = tabela.dropna(how="all", axis=1)
tabela = tabela.dropna(how="any",axis=0)
print(tabela.info())
```

```python
#analizar a real porcentagem de numeros de clientes que cancelaram
print(tabela["Churn"].value_counts())
print(tabela["Churn"].value_counts(normalize="true").map("{:.1%}".format))
```
